```python
from keras.models import load_model
import cv2 as cv
```

```python
model=load_model('SmokingClassification.h5')
def draw_label(img,text,pos,bg_color,text_color):
    text_size=cv.getTextSize(text,cv.FONT_HERSHEY_PLAIN,2,cv.FILLED)
    
    end_x=pos[0]+text_size[0][0]+2
    end_y=pos[1]+text_size[0][1] - 2
    
    cv.rectangle(img,pos,(end_x,end_y),bg_color,cv.FILLED)
    cv.putText(img,text,pos,cv.FONT_HERSHEY_COMPLEX,1,text_color,1,cv.LINE_AA)
```

```python
def smoke_detection(img):
    y_pred=model.predict(img.reshape(1,224,224,3))
    return y_pred[0][0]
```

# For real time detection

```python
capture=cv.VideoCapture(0)
while True:
    ret,frame=capture.read()
    img=cv.resize(frame,(224,224))
    y_pred=smoke_detection(img)
    if y_pred==1:
        draw_label(frame,"Smoking",(20,20),(0,0,255),(0,0,255))
    else:
        draw_label(frame,"No Smoking",(20,20),(255,0,0),(255,0,0))
    cv.imshow('window',frame)
    if cv.waitKey(1) & 0xFF == ord('x'):
        break
cv.destroyAllWindows()
```
